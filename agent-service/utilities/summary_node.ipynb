```python
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
import os
load_dotenv()
llm = ChatOpenAI(
    api_key = os.getenv("OPENROUTER_API_KEY"),
    base_url = "https://openrouter.ai/api/v1",
    model = "google/gemini-2.5-flash",
    # max_completion_tokens=500
)
```

```python
from langgraph.nodes import SummarizationNode

# Initialize the summarization node with desired parameters
summarization_node = SummarizationNode(
    model= llm,
    max_tokens=500,
    max_tokens_before_summary=1000,
    max_summary_tokens=256
)

# Example usage
state = {
    "messages": [
        {"role": "user", "content": "How can I improve my coding skills?"},
        {"role": "assistant", "content": "Practice regularly, read code, and work on projects."},
        # Add more messages as needed
    ]
}

# Process the messages through the summarization node
updated_state = summarization_node.invoke(state)
```
